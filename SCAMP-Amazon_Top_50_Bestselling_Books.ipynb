## Amazon Top 50 Bestselling Books from 2009 - 2019

Analysing the [Kaggle Dataset](https://www.kaggle.com/sootersaalu/amazon-top-50-bestselling-books-2009-2019) with information about the 50 bestselling books on Amazon from 2009 to 2019.

#### _Loading libraries_

```python
%matplotlib inline
import pandas as pd
import matplotlib
import seaborn
```

---

#### _Let's load the books_

```python
file = 'amazon_bestselling_books_09-19.csv'
books_df = pd.read_csv(file)
```

```python
books_df.rename(columns = {'Price':'Price ($)'}, inplace = True)
books_df
```

## 1. The Best Selling Books of the Decade.
---
**a.** Best selling books of the decade according to _user ratings_

**b.** Best selling books of the decade according to _reviews_

### a. Best selling books of the decade according to _user ratings_

```python
maximum_rating =books_df['User Rating'].max()
maximum_rating
```

```python
best_rated_books = books_df[books_df["User Rating"]==maximum_rating].filter(items=['Name', 'Author', 'User Rating'])
best_rated_books
```

### b. Best selling books of the decade according to _reviews_

```python
best_reviewed_books = books_df.sort_values("Reviews", ascending=False).head(10)[['Name', 'Author', 'Reviews']]
best_reviewed_books
```

```python
plot = best_reviewed_books.plot.bar(x='Name', color='#b3066c', figsize=(10,7))
plot.set_title('Best selling books of the decade according to reviews')
plot.set_xlabel('Book Title')
```

## 2. The percentage of fiction to non-fiction genre

```python
genre_df = round(books_df.groupby(books_df['Genre']).count()['Name']/books_df.groupby(books_df['Genre']).count()['Name'].sum() * 100, 2)
plot = genre_df.plot(kind='pie', figsize=(14,8), autopct='%1.1f%%', labels=None, colors=['#b3066c', 'grey'])
plot.set_title('The Percentage of Fiction to Non Fiction Genre')
plot.set_ylabel('Genre')
plot.legend(['Fiction', 'Non Fiction'])
```

## 3. Trends over the decade

---
**a.**  Top 5 Authors who wrote most best selling book    
**b.**  10 best rated books for free     
**c.**  Book with the highest review     
**d.**  The most expensive book  
**e.**  Most Popular Genre over the Decade  

              
 ---

**a. _Top 5 Authors who wrote most best selling books_**

```python
best_selling_authors = books_df.groupby(books_df["Author"]).count()[["Name"]].sort_values("Name", ascending=False).head()
best_selling_authors.rename(columns = {'Name':'No. of times'}, inplace = True)
best_selling_authors
```

```python
plot = best_selling_authors.plot(kind='bar', color=['#b3066c'], figsize=(7,5))
plot.set_title('Top 5 Authors with the most best selling books')
plot.set_xlabel('Best 5 Selling Authors')
```

**b. _10 best rated books for free_**

```python
best_free_books = books_df[books_df["Price ($)"]==0].sort_values("User Rating", ascending=False).head(10)
best_free_books
```

**c. _Book with the highest review_**

```python
highest_reviewed_book = books_df.sort_values("Reviews", ascending=False).head(1)[["Name"]]
highest_reviewed_book
```

**d. _The most expensive book_**

```python
most_expensive_book = books_df.sort_values("Price ($)", ascending=False).head(1)[["Name", "Price ($)"]]
most_expensive_book
```

**e. _Most Popular Genre over the Decade_**

---
The _non fiction genre_ is the most popular genre over the decade

```python
seaborn.countplot(x=books_df['Year'], hue=books_df['Genre'], palette=('PuRd_r'))
```

## 4. Top 10 Authors according to reviews

```python
best_reviewed_authors = books_df.groupby(books_df["Author"]).sum().sort_values("Reviews", ascending=False).head(10).filter(items=['Reviews'])
best_reviewed_authors
```

```python
best_reviewed_authors.plot(kind='bar', color=['#b3066c'], figsize=(10,7))
```
